## 2. Control Flow: conditional statements

“If” uses an expression to evaluate to True or False. If it is True, it executes what it is inside the if statement. For example:

```
if True:
  print("Hello Python If")

if 2 > 1:
  print("2 is greater than 1")
```

The “else” statement will be executed if “if” expression is false.

```
if 1 > 2:
  print("1 is greater than 2")
else:
  print("1 is not greater than 2")
```

You can also use “elif” statement. Like that:



```
if 1 > 2:
  print("1 is greater than 2")
elif 2 > 1:
  print("1 is not greater than 2")
else:
  print("1 is equal to 2")
```

## 3. Looping / Iterator

In Python we can iterate in different forms. We’ll talk about 2: **while** and **for**.


**While Looping: ** while the statement is True, the code inside the block will be executed. So this code will print the number from 1 to 10.

```
num = 1

while num <= 10:
    print(num)
    num += 1
```

The while loop needs a “loop condition” and if it keeps True, it continues iterating. In this example, when num is 11 the “loop condition” evaluate to False.

Another basic code to better understand it:

```
loop_condition = True

while loop_condition:
    print("Loop Condition keeps: %s" %(loop_condition))
    loop_condition = False
```

The “loop condition” is True so it keeps iterating, until we set it to False.


**For Looping:** you pass the variable “num” to the block and the “for” statement will iterate it for you. This code will print the same as “while” code: from 1 to 10.

```

for i in range(1, 11):
  print(i)
```

## 1. List: Collection | Array | Data Structure

Imagine you want to store the integer 1 in a variable. But maybe now you want to store 2. And 3, 4, 5 …


Do I have another way to store all integers that I want, but not in millions variables? 

**List** is a collection that can be used to store a list of values (Like these integers that you want)

```
my_integers = [1, 2, 3, 4, 5]
```

 We created an array and stored it on my_integer.
 
** “How can I get a value from this array?”**

List has a concept called index. The first element gets the index 0 (zero). The second gets 1, and so on. You got the idea!

```
my_integers = [5, 7, 1, 3, 4]
print(my_integers[0]) # 5
print(my_integers[1]) # 7
print(my_integers[4]) # 4
```

Imagine that you don’t want to store integers. You just want to store strings. Like a list of field names.

```
fieldNames = ["Champion","Seria","South West Ampa","Magpie","Mampak","Iron Duke"]

print(fieldNames[4])
```

** “How can I add a value to this array?”**
