```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import lasio
```

Vamos a buscar solo los archivos con formato LAS dentro del directorio ../data/well_logs usando la libreria `glob`, que viene pre-instalada con Python

Vemos que nos retorna una lista con el directorio y el nombre del archivo. Vamos a usar la libreria `lasio` (https://lasio.readthedocs.io/en/latest/index.html), que sirve para leer y escribir archivos con formato LAS (Log ASCII Standard), que es un formato standard usado para datos de registros de pozos.

```python
df1 = w1.df()
```

Ok, vamos ahora a crear un grafico para mostrar nuestro registro de pozo. Mostraremos los siguientes tracks:
- Track 1: GR, CALI
- Track 2: RHOB, NPHI
- Track 3: RM, RACELM, RACEHM, RD

```python
well_logs
```

```python
depths = well_logs.index.values

fig, axs = plt.subplots(nrows=1, ncols=3, sharey=True, squeeze=True, figsize=(15, 10), gridspec_kw={'wspace': 0.25})
fig.subplots_adjust(left=0.05, bottom=0.05, right=0.975, top=0.7, wspace=0.2, hspace=0.2)

axs[0].set_ylabel('Depth (m)')
for ax in axs:
    ax.set_ylim(depths[0],depths[-1])
    ax.invert_yaxis()
    ax.grid(True)
    ax.locator_params(axis='x', nbins=5)
    ax.get_xaxis().set_visible(False)

ax1 = axs[0].twiny()  
ax1.plot(well_logs.GR, well_logs.index, '-', color='olive')
ax1.set_xlim(0,150) 
ax1.set_xlabel('GR (API)', color='olive')
ax1.minorticks_on()
ax1.tick_params(axis='x', color='olive')
ax1.spines['top'].set_position(("axes", 1.15))

ax2 = axs[0].twiny() 
ax2.plot(well_logs.CALI, well_logs.index, '-', color='magenta')
ax2.set_xlim(8,10) 
ax2.set_xlabel('CALI (In)', color='magenta')
ax2.minorticks_on()
ax2.tick_params(axis='x', color='magenta')
ax2.spines['top'].set_position(("axes", 1.075))

ax1 = axs[1].twiny()  
ax1.plot(well_logs.RHOB, well_logs.index, '-', color='blue')
ax1.set_xlim(1.5,3) 
ax1.set_xlabel('RHOB (g/cm3)', color='blue')
ax1.tick_params(axis='x', colors='blue')
ax1.minorticks_on()
ax1.spines['top'].set_position(("axes", 1.075))

ax2 = axs[1].twiny()  
ax2.plot(well_logs.NPHI, well_logs.index, '-', color='forestgreen')
ax2.set_xlim(0.6,-0.1) 
ax2.set_xlabel('NPHI (v/v)', color='forestgreen')
ax2.tick_params(axis='x', color='forestgreen')
ax2.minorticks_on()
ax2.spines['top'].set_position(("axes", 1.0))


ax1 = axs[2].twiny()  
ax1.plot(well_logs.RACEHM, well_logs.index, '-', color='lime')
ax1.grid(True)
ax1.set_xlim(0.1,2000)  
ax1.set_xlabel('RACEHM (ohm.m)', color='lime')
ax1.tick_params(axis='x', color='lime')
ax1.set_xscale('log')
ax1.minorticks_on()
ax1.spines['top'].set_position(("axes", 1.075))

ax2 = axs[2].twiny()  
ax2.plot(well_logs.RACELM, well_logs.index, '-', color='darkviolet')
ax2.set_xlim(0.1,2000) 
ax2.set_xlabel('RACELM (ohm.m)', color='darkviolet')
ax2.tick_params(axis='x', colors='darkviolet')
ax2.set_xscale('log')
ax2.minorticks_on()
ax2.spines['top'].set_position(("axes", 1.15))

ax3 = axs[2].twiny()  
ax3.plot(well_logs.RM, well_logs.index, '-', color='darkblue')
ax3.set_xlim(0.1,2000)  
ax3.set_xlabel('RM (ohm.m)', color='darkblue')
ax3.tick_params(axis='x', color='darkblue')
ax3.set_xscale('log')
ax3.minorticks_on()
ax3.spines['top'].set_position(("axes", 1.225))

ax4 = axs[2].twiny()  
ax4.plot(well_logs.RD, well_logs.index, '-', color='cyan')
ax4.set_xlim(0.1,2000) 
ax4.set_xlabel('RD (ohm.m)', color='cyan')
ax4.tick_params(axis='x', color='cyan')
ax4.set_xscale('log')
ax4.minorticks_on()
ax4.spines['top'].set_position(("axes", 1.0))

fig.suptitle(f'15_9-F-11A',weight='bold', fontsize=20, y=0.925)
```

Ahora pongamos todo este codigo en una funcion, asi podemos hacer plots de todos los pozos sin repetir el codigo.

Este codigo es para extraer el nombre del pozo usando Lasio

```python
w1.header['Well']['WELL']['value']
```

```python
def plot_well_logs(filename):
    # Obtener nombre del pozo:
    well = filename.split("\\")[-1].split('.LAS')[0]
    well = lasio.read(filename)
    well_name = well.header['Well']["WELL"]['value']
    df = well.df()
    df = df[["GR","CALI","RHOB","NPHI","RACELM","RACEHM"]]
    well_logs = df.dropna()
    depths = well_logs.index.values

    fig, axs = plt.subplots(nrows=1, ncols=3, sharey=True, squeeze=True, figsize=(15, 10), gridspec_kw={'wspace': 0.25})
    fig.subplots_adjust(left=0.05, bottom=0.05, right=0.975, top=0.7, wspace=0.2, hspace=0.2)

    axs[0].set_ylabel('Depth (m)')
    for ax in axs:
        ax.set_ylim(depths[0],depths[-1])
        ax.invert_yaxis()
        ax.grid(True)
        ax.locator_params(axis='x', nbins=5)
        ax.get_xaxis().set_visible(False)

    ax1 = axs[0].twiny()  
    ax1.plot(well_logs.GR, well_logs.index, '-', color='olive')
    ax1.set_xlim(0,150) 
    ax1.set_xlabel('GR (API)', color='olive')
    ax1.minorticks_on()
    ax1.tick_params(axis='x', color='olive')
    ax1.spines['top'].set_position(("axes", 1.15))

    ax2 = axs[0].twiny() 
    ax2.plot(well_logs.CALI, well_logs.index, '-', color='magenta')
    ax2.set_xlim(8,10) 
    ax2.set_xlabel('CALI (In)', color='magenta')
    ax2.minorticks_on()
    ax2.tick_params(axis='x', color='magenta')
    ax2.spines['top'].set_position(("axes", 1.075))

    ax1 = axs[1].twiny()  
    ax1.plot(well_logs.RHOB, well_logs.index, '-', color='blue')
    ax1.set_xlim(1.5,3) 
    ax1.set_xlabel('RHOB (g/cm3)', color='blue')
    ax1.tick_params(axis='x', colors='blue')
    ax1.minorticks_on()
    ax1.spines['top'].set_position(("axes", 1.075))

    ax2 = axs[1].twiny()  
    ax2.plot(well_logs.NPHI, well_logs.index, '-', color='forestgreen')
    ax2.set_xlim(0.6,-0.1) 
    ax2.set_xlabel('NPHI (v/v)', color='forestgreen')
    ax2.tick_params(axis='x', color='forestgreen')
    ax2.minorticks_on()
    ax2.spines['top'].set_position(("axes", 1.0))


    ax1 = axs[2].twiny()  
    ax1.plot(well_logs.RACEHM, well_logs.index, '-', color='lime')
    ax1.grid(True)
    ax1.set_xlim(0.1,2000)  
    ax1.set_xlabel('RACEHM (ohm.m)', color='lime')
    ax1.tick_params(axis='x', color='lime')
    ax1.set_xscale('log')
    ax1.minorticks_on()
    ax1.spines['top'].set_position(("axes", 1.075))

    ax2 = axs[2].twiny()  
    ax2.plot(well_logs.RACELM, well_logs.index, '-', color='darkviolet')
    ax2.set_xlim(0.1,2000) 
    ax2.set_xlabel('RACELM (ohm.m)', color='darkviolet')
    ax2.tick_params(axis='x', colors='darkviolet')
    ax2.set_xscale('log')
    ax2.minorticks_on()
    ax2.spines['top'].set_position(("axes", 1.15))

    fig.suptitle(f'{well_name}',weight='bold', fontsize=20, y=0.925)    
    
```

```python
plot_well_logs(all_las[0])
```

Ahora hagamos plots para todos los pozos!

```python
for w in all_las:
    plot_well_logs(w)
```
